### 유튜브 크롤링
- 타이틀, 조회 수, 시간 수집

```python
import time
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from bs4 import BeautifulSoup as bs

# 0. setting
driver = webdriver.Chrome()
url = "https://www.youtube.com/@kimfl/videos"
driver.get(url)
```

```python
# 1. 스크롤
# 자바스크립트로 내리기, body+end키
while True:
    prevHeight = driver.execute_script("return document.documentElement.scrollHeight")

    driver.execute_script("window.scrollTo(0, document.documentElement.scrollHeight)")
    time.sleep(2)
    
    nowHeight = driver.execute_script("return document.documentElement.scrollHeight")
    # 아무리 스크롤을 내려도 전체 길이가 변화가 없다면 모든 컨텐츠가 로딩된 것
    
    if prevHeight == nowHeight:
        break
```

```python
# 2. 데이터 수집
import pandas as pd

soup = bs(driver.page_source, "html.parser")
boxs = soup.select("ytd-rich-grid-media.ytd-rich-item-renderer")


titleSelector = "yt-formatted-string.style-scope.ytd-rich-grid-media#video-title"
reviewSelector = "#metadata-line"

titles = []
views = []
times = []

for box in boxs:
    # title
    title = box.select_one(titleSelector).text
    titles.append(title)

    # view, time
    viewAndTime = box.select_one(reviewSelector).text
    tmp = viewAndTime.split("\n")
    views.append(tmp[3][4:])
    times.append(tmp[4])

data = {"제목": titles, "조회수": views, "등록일": times}

result = pd.DataFrame(data)
result
result.to_csv("유튜브크롤링.csv", encoding="utf-8-sig")
```
