- 스크립트 언어 : 문자열 그대로 실행되는 프로그래밍 언어

```python
import requests as req
from bs4 import BeautifulSoup as bs

query = input("무엇을 검색하시겠습니까?")
url = f"https://search.danawa.com/dsearch.php?query={query}"
headers = {'User-Agent': 'Mozilla/5.0'}
res = req.get(url=url, headers=headers)
soup = bs(res.text, "html.parser")
# print(soup.prettify())
```

```python
items = soup.select(".prod_main_info")

for item in items:
    # title
    titleNode = item.select_one(".click_log_product_standard_title_")  # select_one -> 첫번째 결과 하나 반환
    title = titleNode.text if titleNode != None else "타이틀 없음"

    # price
    priceNode = item.select_one(".rank_one>.price_sect>.click_log_product_standard_price_")
    price = priceNode.text if priceNode != None else 0
    
    price = price.replace(",", "")
    price = price.replace("원", "")
    price_int = int(price)
    
    # point
    pointNode = item.select_one(".point_num")
    point = pointNode.text if pointNode != None else "리뷰없음"

    print(f'{title} / {price}원 / {point}')
```
