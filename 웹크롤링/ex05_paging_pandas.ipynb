### paging 크롤링

1. 다나와에서 검색하여 10페이지의 검색 정보를 가져온다.
2. pandas를 활용해 표 형식으로 만든다.

```python
import requests as req
import pyautogui as pg# gui 키보드와 마우스를 자동으로 움직이는 라이브러리 + 창(alert, prompt, password, confirm)
from bs4 import BeautifulSoup as bs

query = pg.prompt("검색어를 입력하세요")  # 맥에서 종료가 안되는 오류 발생

# 리스트
titles = []
prices = []
pointNums = []
pointCnts = []


for page in range(1, 11):

    url = f"https://search.danawa.com/dsearch.php?query={query}&{page}=2&limit=40"
    headers = {'User-Agent': 'Mozilla/5.0'}
    res = req.get(url=url, headers=headers)
    soup = bs(res.text, "html.parser")  # http통신이 현재는 모든 파일을 주고 받을 수 있다
    print(f'=================================={page}페이지=====================================')
    prods = soup.select(".prod_main_info")  # 검색한 아이템 정보

    
    for prod in prods:
        # title
        titleNode = prod.select_one(".click_log_product_standard_title_") 
        title = titleNode.text if titleNode != None else "제목 없음"

        # price
        priceNode = prod.select_one(".click_log_product_standard_price_ > strong")
        price = priceNode.text if priceNode != None else 0
        
        price = price.replace(",", "")
        price = price.replace("원", "")
        price_int = int(price)

        # point
        pointNumNode = prod.select_one(".point_num >strong")
        pointNum = float(pointNumNode.text) if pointNumNode != None else "평점 없음"

        pointCntNode = prod.select_one(".click_log_prod_content_count > strong")
        pointCnt = pointCntNode.text if pointCntNode != None else "리뷰 없음"

        # print(f'{title} / {price}원 / {pointNum}점 / {pointCnt}건')
        
        # 리스트 만들기
        titles.append(title)
        prices.append(price_int)
        pointNums.append(pointNum)
        pointCnts.append(pointCnt)
```

```python
import pandas as pd

data = {"상품명": titles, "가격": prices, "평점": pointNums, "리뷰": pointCnts}


result = pd.DataFrame(data=data)
result
```
